```python
import pandas
import sqlite3
import numpy
from datetime import date, timedelta
```

```python

# how many steps left for the rest of the year
def find_new_yearly() -> int:
    conn = sqlite3.connect('count_steps.db')
    cursor = conn.cursor()

    total = cursor.execute('SELECT SUM(Count) FROM steps_by_date').fetchone()[0]
    conn.commit()

    conn.close()
    return 5000000 - total

# find the new daily average
def find_new_daily() -> int:
    conn = sqlite3.connect('count_steps.db')
    cursor = conn.cursor()

    day_count = cursor.execute('SELECT COUNT(*) FROM steps_by_date').fetchone()[0]
    conn.commit()

    conn.close()

    new_daily = round(find_new_yearly() / (365 - day_count))
    return new_daily

#add entries to the sqlite db
def add_or_update_entry(date: int, count: int) -> None:
    conn = sqlite3.connect('count_steps.db')
    cursor = conn.cursor()

    cursor.execute('INSERT OR REPLACE INTO steps_by_date(Date, Count) VALUES(?, ?)', (date, count))
    conn.commit()

    conn.close()
```
